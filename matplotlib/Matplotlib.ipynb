```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

## Functional Plot

```python
x = np.linspace(0, 10, 20)
y = x ** 2
plt.plot(x, y)
plt.xlabel('Days')
plt.ylabel('Squared')
plt.title('Squared Chart')
plt.show()
```

## Print Multiple Plots

```python
plt.subplot(2, 2, 1) # Row, Col, ID
plt.plot(x, y, 'r')
plt.subplot(2, 2, 2)
plt.plot(y, x, 'g')
plt.subplot(2, 2, 3)
plt.plot(x, x, 'black')
plt.subplot(2, 2, 4)
plt.plot(x, np.sin(y), 'cyan')
plt.show()
```

## Using Figure Objects

```python
fig_1 = plt.figure(figsize=(5, 4), dpi=100) # dpi for the resolution of the plot
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9]) # array contains how much space is taken in figure
axes_1.plot(x, y, label='x/x^2')
axes_1.plot(y, x, label='x^2/x')
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Squared')
axes_1.set_title('Squared Chart')
# 1: upper right, 2: upper left, 3: lower right, 4: lower left
# or loc=(x, y) from lower left
axes_1.legend(loc=0)

# Creating another axes inside axes_1
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Squared')
axes_2.set_title('Days Squared Chart')
axes_2.plot(x, x, 'r')

axes_2.text(0, 15, 'Custom Text') # .text(left, right, msg)

plt.show()
```

## SubPlots

```python
fig, ax = plt.subplots(figsize=(8, 4), nrows=1, ncols=3, sharey=True)
plt.tight_layout() # So that the plots don't overlap
ax[1].plot(x, y, label='Squared', color='m')
ax[1].legend(loc=0)
ax[0].plot(y, x, label='Inverted', color='r')
ax[0].legend(loc=0)
ax[2].plot(x, np.cos(x), label='cos', color='g')
ax[2].legend(loc=0)
plt.show()
```

## Appearance Options

```python
fig_2 = plt.figure(figsize=(6, 4))
axes_2 = fig_2.add_axes([0, 0, 1, 1])
axes_2.plot(x, y, color='r', linestyle='-.', linewidth=2, 
            marker='o', markersize=7, markerfacecolor='y',
            markeredgecolor='y', markeredgewidth=3,
           alpha=0.75)

# Zoom at particular interval in the graph
axes_2.set_xlim([0, 3])
axes_2.set_ylim([0, 8])

axes_2.grid(True, color='0.6', dashes=(5, 2, 1, 2)) # Show grid
axes_2.set_facecolor('cyan') # Backgroud color
```

## Save Visualization to a File

```python
fig_2.savefig('SamplePlot.png')
```

## Pie Charts

```python
import random

fig_5 = plt.figure(figsize=(7, 4), dpi=100)
axes_5 = fig_5.add_axes([0.1, 0.1, 0.9, 0.9])

types = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
num = [100, 120, 50, 40, 70, 150, 20]

colors = ['navy', 'red', 'blue', 'black', 'green', 'magenta', 'orange']
explode = [0] * 7
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(num, labels=types, explode=explode, 
                                   colors=colors, shadow=True, startangle=140, 
                                   autopct='%1.0f%%', textprops=dict(color='w'))

plt.legend(wedges, types, loc='right', bbox_to_anchor=(1, 0, 0.2, 1))
plt.show()
```

## 3D Surface

```python
from mpl_toolkits import mplot3d

fig_8 = plt.figure(figsize=(8, 7), dpi=100)
axes_8 = fig_8.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')

np.random.seed(10)
z = 20 * np.random.randint(0, 100, 100)
x = np.sin(z) * np.random.randn(100)
y = np.tan(z) * np.random.randn(100)
axes_8.scatter3D(x, y, z, c=z, cmap='Blues')
plt.show()
```

```python
def get_z(x, y):
    return np.sin(np.tan(x ** 2 + y ** 2))

x = np.linspace(-10, 10, 50)
y = np.linspace(-10, 10, 50)

x, y = np.meshgrid(x, y)
z = get_z(x, y)

fig_9 = plt.figure(figsize=(10, 8), dpi=100)
axes_9 = fig_9.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')

axes_9.view_init(25, 55)
axes_9.contour3D(x, y, z, 200, cmap='rainbow')
plt.show()
```
